```python
import json

input_path = r"C:\Datasets\processed\time_forge_ver_2_syn.jsonl"
output_path = r"C:\Datasets\processed\clean_time_forge.jsonl"

seen = set()
clean_data = []

with open(input_path, "r", encoding="utf-8") as f:
    for line in f:
        data = json.loads(line)
        
        key = data["instruction"] + data["response"]
        
        if key not in seen:
            seen.add(key)
            clean_data.append(data)

with open(output_path, "w", encoding="utf-8") as f:
    for item in clean_data:
        f.write(json.dumps(item) + "\n")

print("Original:", len(seen))
print("Cleaned:", len(clean_data))
```

```python
from collections import Counter
import json

topic_count = Counter()

with open(r"C:\Datasets\processed\clean_time_forge.jsonl", "r") as f:
    for line in f:
        data = json.loads(line)
        topic = data["instruction"].split("for")[-1].strip()
        topic_count[topic] += 1

print(topic_count.most_common(10))
```
